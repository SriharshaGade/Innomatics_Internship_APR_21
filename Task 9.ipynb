```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.stats import t
```

### Q-1 Pista House selling Hyderabadi Chicken Dum biryani claims that each parcel packet has 500 grams of biryani. You are sceptic of their claims and believe that on average each packet does not contain 500 grams of biryani. How do you prove your claim?

<h3>
    <font color='brown'>Step 1:</font><p>&nbsp;</p>
    Null Hypothesis ($H_0$) : The weight of the biryani packet is grater than or eqaul to 500g.<p>&nbsp;</p>
    $$H_0 : \bar{x}=500$$<br>
    Alternate Hypothesis ($H_1$) : The weight of the biryani packet is less than 500g.<p>&nbsp;</p>
    $$H_1: \bar{x}\neq500 $$
</h3>

<h3>
<font color='brown'>Step 2: Sample Collection</font>
</h3>

```python
pop_mean = 500 #Given
sample = [490,220,470,500,495,496,496,498,508,480]
N = len(sample)
sample_mean = sum(sample)/N
print("The Mean for Given Observations is {}".format(sample_mean))
```

<h3>
<font color='brown'>Step 3:</font><p>&nbsp;</p>
$$t=\frac{\bar{x}-\mu}{{s}/{\sqrt{N}}}$$
</h3>

<h3><i>Where,<br>t = Test value<br>$\bar{X}$ = Mean of the sample<br>$\mu$ = Mean of the population<br>s = standard deviation of the sample<br>N = Sample size<br></i></h3>

```python
def t_score(sample_size, sample_mean, pop_mean, sample_std):
    numerator = sample_mean - pop_mean
    denomenator = sample_std / sample_size**0.5
    return numerator / denomenator
```

```python
sample_std = np.std(sample)
print("The Standard Deviation is {}". format(sample_std))
```

```python
t_val = t_score(N,sample_mean,pop_mean,sample_std)
print("The T static value is {}".format(t_val))
```

<h3>
<font color='brown'>Step 4:</font>
</h3>

## The confidence level is 95% i.e. 0.95 <br> $\alpha$ = 1 - confidence level <br> Therefore $\alpha$ = 0.05

```python
confidence_interval = 0.95
alpha = 1 - confidence_interval
```

```python
t_critical = t.ppf(1-alpha/2,df = N-1)
print("The T critical value is {}".format(t_critical))
```

```python
if abs(t_val) >= t_critical:
	print("Reject the Null Hypothesis.")
else:
	print('Fail to Reject the Null Hypothesis.')
```

<h3>
<font color='brown'>Step 5:</font>
</h3>

## Now Plot the T-critical value and T static value to plot out the hypothesis output.

```python
x_min = 400
x_max = 600
df = N-1

mean = pop_mean
std = sample_std


x = np.linspace(x_min, x_max, 100)
y = t.pdf(x, df,mean, std)
plt.figure(figsize = (15,10))


t_critical_left = pop_mean - (t_critical * std)
t_critical_right = pop_mean + (t_critical * std)


x1 = np.linspace(x_min, t_critical_left, 100)
y1 = t.pdf(x1,df, mean, std)
plt.fill_between(x1, y1, color='red')
plt.scatter(x1,y1)


x2 = np.linspace(t_critical_right, x_max, 100)
y2 = t.pdf(x2,df, mean, std)
plt.fill_between(x2, y2, color='red')
plt.scatter(x2,y2)
plt.scatter(sample_mean, 0)
plt.annotate("Sample_mean", (sample_mean-20, 0.0007))
plt.axvline(mean, color='green')
plt.annotate("pop_mean", (mean+1, 0.01))


plt.plot(x, y)
plt.show()
```

## Failed to Reject Null Hypothesis and Validate the average Mean of biryani Packet is 500gms.

### Q-4 You have developed a new Machine Learning Application and claim that on average it takes less than 100 ms to predict for any future datapoint. How do you convince your client about this claim?

<h3>
<font color='brown'>Step 3:</font><p>&nbsp;</p>
$$t=\frac{\bar{x}-\mu}{{s}/{\sqrt{N}}}$$
